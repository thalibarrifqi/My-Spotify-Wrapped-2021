```python
def main(time):
    tracks = sp.current_user_top_tracks(time_range=time, limit=50)
    track_data = []
    for i, item in enumerate(tracks['items']):
        trackid = item['id']
        trackurl = item['external_urls']['spotify']
        artwork = item['album']['images'][0]['url']
        title = item['name']
        artist = item['artists'][0]['name']
        properties = {
            'trackid':trackid,
            'trackurl':trackurl,
            'artwork':artwork,
            'title':title,
            'artist':artist
        }
        track_data.append(properties)

    name_date = datetime.now().strftime("%m%d%Y%H%M%S")
    
    #save to csv
    df_track = pd.DataFrame(track_data)
    df_track.to_csv(time+"-"+name_date+".csv", sep = ',', index=False)

    #store to spreadsheet
    worksheet = sh.worksheet(time)
    val = worksheet.acell('B5').value
    worksheet.update([df_track.columns.values.tolist()] + df_track.values.tolist())
```

```python
times = ['long_term', 'medium_term', 'short_term']
```

```python
for time in times:
    main(time)
```
